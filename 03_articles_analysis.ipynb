### Analisis de datos
Saber cuales de los diferentes sitios web tienen mas interaccion

```python
#seleccionamos columnas bases para el analisis
COLUMNS = ['source_name',
           'author',
           'engagement_reaction_count',
           'engagement_comment_count',
           'engagement_share_count',
           'engagement_comment_plugin_count'
          ]

#limpiar todas aquellas fillas que tienen valor nulo en columnas seleccionadas
data = data[COLUMNS].dropna(axis=0,how='any')
```

```python
#Agrupamos nuestro data frame por la com¿lumna source_name y sumamos cada fila por su valor en la columna en 	engagement_reaction_count
engagement_reaction_count  = data.groupby('source_name')['engagement_reaction_count'].sum()
engagement_reaction_count.to_frame()
```

```python
engagement_share_count  = data.groupby('source_name')['engagement_share_count'].sum()
engagement_share_count.to_frame()
```

```python
#estraer nueva serie 

highest_source = data['source_name'].value_counts()
type(highest_source)

```

```python
#estraer nueva serie (2) 
highest_source.to_frame()
```

### Visualizacion de datos

```python
# seleccionamos el valor de la primera columna (indice) u lo convertimos a un elemento de lista por cada valor en la serie
highest_source_labels = list(highest_source.index.values)
print(highest_source_labels)
```

```python
#crear grafica de pastel

plt.pie(highest_source,
        labels = highest_source_labels,
        autopct='%1.1f%%')

plt.title('Fuentes de informacion')
plt.show()
```
